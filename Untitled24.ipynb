# DAX Functions in Power BI

1 . Create a calculated column TotalPrice using Quantity and Unit Price.

Sol :

 DAX

     TotalPrice =

       [Quantity] * [UnitPrice]


2 . Create a calculated column DiscountedPrice.

Sol :       

DiscountedPrice =

[TotalPrice] - ([TotalPrice] * [Discount] / 100)

3 . Create a measure Average Order Value (AOV).

Sol :      

 AOV =
DIVIDE(
    SUM('Sales'[SalesAmount]),
    DISTINCTCOUNT('Sales'[OrderID])
)

4 . Using DAX, create a measure that calculates cumulative Year-to-Date (YTD) Sales based on the Order
Date

Sol :        

 YTD Sales =
TOTALYTD(
    SUM('Sales'[SalesAmount]),
    'Sales'[OrderDate]
)

5 . Using DAX, create a measure to rank each product according to its Total Sales, with the highest-selling
product ranked first.

Sol :      

  Product Rank =

RANKX(

    ALL('Sales'[ProductID]),
    CALCULATE(SUM('Sales'[SalesAmount])),
    ,
    DESC,
    DENSE
)

6 . Create a calculated column classifying orders as “High Value” if sales > 2000, else “Low Value”.

Sol :       
 Order Category =

IF(

    [SalesAmount] > 2000,
    "High Value",
    "Low Value"
)

7 . Create a measure to calculate total sales from the Online channel only

Sol :       
 Online Sales =

CALCULATE(

    SUM('Sales'[SalesAmount]),
    'Sales'[SalesChannel] = "Online"
)


8 . Create a measure called "Date Difference" that calculates the number of days between the latest  Order
Date and today's date using the DATEDIFF function.

Sol :    

 Date Difference =

DATEDIFF(
  
    MAX('Sales'[OrderDate]),
    TODAY(),
    DAY
)
